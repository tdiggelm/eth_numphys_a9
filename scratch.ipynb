```python
# non-symmetric

a, b = -10.0, 10.0
v = lambda x: 0.5*x**2
NN = 2**arange(5,11)
EW = []
EV = []
for N in NN:
    x, h = linspace(a, b, N, retstep=True)

    L = diag(N*[-2])+eye(N,k=1)+eye(N,k=-1)
    L[0]=L[1]
    L[-1]=L[-2]
    L = 1/h**2*L

    H = -0.5*L+diag(v(x))
    ew, ev = eig(H)
    
    ew = abs(ew)
    ev = ev.T.real
    
    ew, ev = zip(*[(ew[i], ev[i]) for i in argsort(ew)])
    #ew, ev = zip(*sorted(zip(ew, ev),key=lambda x: x[0])[:nn])
    EW.append(ew)
    EV.append(ev)
```

```python
# symmetric
a, b = -10.0, 10.0
NN = 2**arange(5,11)
EW = []
EV = []
for N in NN:
    x, h = linspace(a, b, N, retstep=True)
    v = lambda x: 0.5*x**2

    L = diag(N*[-2])+eye(N,k=1)+eye(N,k=-1)
    L = 1/h**2*L

    H = -0.5*L+diag(v(x))
    ew, ev = eig(H)
    
    ew = abs(ew)
    ev = ev.T.real
    
    ew, ev = zip(*[(ew[i], ev[i]) for i in argsort(ew)])
    #ew, ev = zip(*sorted(zip(ew, ev),key=lambda x: x[0])[:nn])
    EW.append(ew)
    EV.append(ev)
```

```python
def arnoldi(A, v0, k):
    r"""Arnoldi algorithm to compute the Krylov approximation :math:`H` of a matrix :math:`A`.

    :param A: The matrix :math:`A` of shape :math:`N \times N` to approximate.
    :param v0: The initial vector :math:`v_0` of length :math:`N`.
    :param k: The number :math:`k` of Krylov steps performed.
    :return: A tuple :math:`(V, H)` where :math:`V` is the large matrix of shape
             :math:`N \times (k+1)` containing the orthogonal vectors and :math:`H` is the
             small matrix of shape :math:`k \times k` containing the Krylov approximation
             of :math:`A`.
    """
    r, c = A.shape
    V = zeros((r, k+1), dtype=complexfloating)
    H = zeros((k+1, k), dtype=complexfloating)

    ########################################################
    #                                                      #
    # TODO: Implementieren Sie hier das Arnoldi Verfahren. #
    #                                                      #
    ########################################################
    V[:,0] = v0/norm(v0) # V[:,0] = v0.copy/norm(v0)
    for m in xrange(k):
        vt = A.dot(V[:,m])
        for j in xrange(m+1):
            H[j,m] = dot(V[:,j].conjugate(), vt)
            vt -= H[j,m] * V[:,j]
        H[m+1,m] = norm(vt)
        V[:,m+1] = vt/H[m+1,m]
    return V[:,:-1], H[:-1,:]

def hamiltonian(V, N, t0, te, symmetric=False):
    x, h = linspace(t0, te, N, retstep=True)
    L = diag(N*[-2])+eye(N,k=1)+eye(N,k=-1)
    if not symmetric:
        L[0]=L[1]
        L[-1]=L[-2]
    L = 1/h**2*L
    H = -0.5*L+diag(v(x))
    return (x, H)

N = 265
d = 16
a = 1
v = lambda x: d*(exp(-2*a*x) - 2*exp(-a*x))
x, H = hamiltonian(v, N, t0, te, symmetric=True)

idx = 0

ew, ev = eig(H)
ew = abs(ew)
ev = ev.T.real
ew, ev = zip(*[(ew[i], ev[i]) for i in argsort(ew)])
psi = ev[idx]

v0 = rand(265)
V, HH = arnoldi(H, v0, 150)
ew, ev = eig(HH)
ew = abs(ew)
ev = ev.T.real
ew, ev = zip(*[(ew[i], ev[i]) for i in argsort(ew)])
psi_a = V.dot(ev[idx])

psi = abs(psi)
psi_a = abs(psi_a)

plot(x, psi)
plot(x, psi_a)
```
