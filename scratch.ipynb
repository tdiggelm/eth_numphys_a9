```python
from numpy import *
from scipy import *
from scipy.linalg import *
from matplotlib.pylab import *

%matplotlib inline
```

```python
# non-symmetric

a, b = -10.0, 10.0
v = lambda x: 0.5*x**2
NN = 2**arange(5,11)
EW = []
EV = []
for N in NN:
    x, h = linspace(a, b, N, retstep=True)

    L = diag(N*[-2])+eye(N,k=1)+eye(N,k=-1)
    L[0]=L[1]
    L[-1]=L[-2]
    L = 1/h**2*L

    H = -0.5*L+diag(v(x))
    ew, ev = eig(H)
    
    ew = abs(ew)
    ev = ev.T.real
    
    ew, ev = zip(*[(ew[i], ev[i]) for i in argsort(ew)])
    #ew, ev = zip(*sorted(zip(ew, ev),key=lambda x: x[0])[:nn])
    EW.append(ew)
    EV.append(ev)
```

```python
# symmetric
a, b = -10.0, 10.0
NN = 2**arange(5,11)
EW = []
EV = []
for N in NN:
    x, h = linspace(a, b, N, retstep=True)
    v = lambda x: 0.5*x**2

    L = diag(N*[-2])+eye(N,k=1)+eye(N,k=-1)
    L = 1/h**2*L

    H = -0.5*L+diag(v(x))
    ew, ev = eig(H)
    
    ew = abs(ew)
    ev = ev.T.real
    
    ew, ev = zip(*[(ew[i], ev[i]) for i in argsort(ew)])
    #ew, ev = zip(*sorted(zip(ew, ev),key=lambda x: x[0])[:nn])
    EW.append(ew)
    EV.append(ev)
```

```python
nn = 32
figure(figsize=(9,7))
for N, E, Psi in zip(NN, EW, EV):
    plot(E[:nn], label=r"$N=%d$" % N)
    grid(True)
    xlim(0, nn)
    xlabel(r"$n$")
    ylabel(r"$E_n^{(N)}$")
    legend(loc='best')
show()
```

```python
figure(figsize=(9,7))
Ee = arange(nn)+0.5
for N, E, Psi in zip(NN, EW, EV):
    plot(abs(E[:32]-Ee), label=r"$N=%d$" % N)
    grid(True)
    xlim(0, nn)
    xlabel(r"$n$")
    ylabel(r"$|E_n^{(N)}-E_n^{exact}|$")
    legend(loc='best')
show()
```

```python
from scipy.special import hermite, gamma
def psi_exact(n, x):
    C = 1.0 / sqrt(2**n * gamma(n+1)) / sqrt(sqrt(pi))
    return C * exp(-0.5*x**2) * polyval(hermite(n), x)

x = linspace(a, b, 1024)
N, E, psi = NN[-1], EW[-1], EV[-1]
```

```python
psi_e = array([psi_exact(i,x) for i in range(7)])
psi_e = (psi_e.T / norm(psi_e, axis=1)).T

figure(figsize=(9,7))
for n in range(7):
    plot(x, abs(psi_e[n])-abs(psi[n]), label="$n=%d$" % n)
grid(True)
legend(loc='lower right')
xlim(-6, 6)
#ylim(1e-10, 1e-3)
xlabel(r"$x$")
ylabel(r"$\||\psi_n^{exakt}|-|\psi_n^{(N)}||$")
```
